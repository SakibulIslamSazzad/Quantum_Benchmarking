```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import cv2
```

```python
# create figure
fig = plt.figure(figsize=(10, 18))
  
# setting values to rows and column variables
rows = 3
columns = 1
  
# reading images
Image1 = cv2.imread('Images/Bar_charts/bit_error_GHZ.jpg')
Image2 = cv2.imread('Images/Bar_charts/depolar_GHZ.jpg')
Image3 = cv2.imread('Images/Bar_charts/t1_t2_GHZ.jpg')

  
# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)
  
# showing image
plt.imshow(Image1)
plt.axis('off')
plt.title("GHZ")
  
# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)
  
# showing image
plt.imshow(Image2)
plt.axis('off')
plt.title("Mermin")
  
# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)
  
# showing image
plt.imshow(Image3)
plt.axis('off')
plt.title("Phase")

plt.savefig('try.pdf')
```
